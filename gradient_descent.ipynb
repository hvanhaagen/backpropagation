```python
import numpy as np
import matplotlib.pyplot as plt
```

Basic gradient descent on a quadratic function

```python
x_start = 8
lr = 0.1 # learning rate
aantal_iteraties = 20

x_iter = []
y_iter = []

for m in range(aantal_iteraties):
    
    error = 2*x_start
    x_start = x_start - error*lr # de update rule
     
    x_iter += [x_start]
    y_iter += [x_start**2]
```

Plot The quadratic function. In red you see gradient dots. 

```python
x = np.linspace(-5, 5, num=100)
y = x**2

plt.axis([-4, 4, -1, 10])
plt.grid()
plt.plot(x,y)

plt.plot(x_iter,y_iter,'o',color='red')
plt.title('Gradient Descent')
plt.show()
```
