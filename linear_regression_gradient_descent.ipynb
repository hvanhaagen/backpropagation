```python
import numpy as np
import matplotlib.pyplot as plt
```

We make a linear line $y=ax+b$ and add some noise to make it look realistic. We are going to train a regression line. 

```python
noise = np.random.normal(0,1,100)*0.5

x = np.random.normal(0,1,100)
y = 2*x+3 + noise
```

As you can see the slope is a=2 and the intercept is b=3. 

```python
plt.scatter(x,y)
```

```python
lr = 0.01 # learning rate
epoch = 10
```

```python
# start values for a and b. Those are estimation and that is why we call them hat. 
ahat = 10
bhat = 10
# in the end ahat should become 2, and bhat should become 3
```

```python
monitor_a = []
monitor_b = []
```

So we see the terms like epoch, learning rate, and the update rule. The most import thing is calculate the gradient using the chainrule. 

We start with estimations for the slope $\hat{a}$ and intercept $\hat{b}$. We then get the line $\hat{y}=\hat{a}x+\hat{b}$. The error is defined as $(y-\hat{y})^2$. By taking the derivative and using the chainrule we get the gradients for the slope and for the intercept. 

```python
for n in range(epoch):
    
    N = len(x)
    for k in range(N):
        
        yhat = ahat*x[k]+bhat
    
        error = yhat-y[k] # derivative of error
        
        delta_x = error*x[k] # x[k] is derivative with respect to a
        delta_b = error*1  # 1 is derivative with respect to b
    
        ahat = ahat - lr*delta_x # update rule
        bhat = bhat - lr*delta_b # update rule

        monitor_a += [ahat] # keep track of the learning proces
        monitor_b += [bhat]
```

After training we get a beautiful regression line

```python
x_r = np.linspace(-3, 3, num=100)
y_r = ahat*x_r+bhat

plt.scatter(x,y)
plt.scatter(x_r,y_r,color='red')
```

```python
# We check what the final results are for a and b. 
# Indeed a comes close to 2. And be comes close to 3. 
print(ahat)
print(bhat)
```

```python
# We keep track of the learning. We see that ahat starts at 10 and becomes 2. 
plt.plot(monitor_a)
```

```python
# the same for b. It starts at 10 and becomes 3. 
plt.plot(monitor_b)
```
