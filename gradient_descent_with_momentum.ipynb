```python
import numpy as np
import matplotlib.pyplot as plt
```

Here we do gradient descent on a more complex function with two minima. One minimum is local and the other one the global minimum. With just gradient descent we get stuck in the local minimum

```python
x_start = 3.5
lr = 0.01 # learning rate
number_of_iteration = 50

# keep track of process
x_iter = []
y_iter = []

# The original function: x**2+np.sin(2*x)*3
# The derivative: 2*x+np.cos(2*x)*6

for m in range(number_of_iteration):
    
    error = 2*x_start+np.cos(2*x_start)*6
    
    x_start = x_start - error*lr # de update rule
     
    x_iter += [x_start]
    y_iter += [x_start**2+np.sin(2*x_start)*3]
```

```python
# We plot the function in the range of -2 and 5. 
x = np.linspace(-2, 5, num=100)
y = x**2+np.sin(2*x)*3

plt.axis([-2, 4, -3, 15])
plt.grid()
plt.plot(x,y)
# We plot the values produced by gradient descent
plt.plot(x_iter,y_iter,'o',color='red')
plt.title("Gradient descent stuck in a local minimum")
plt.show()
```

```python
x_start = 3.5
lr = 0.01 # learning rate
number_of_iteration = 50

# keep track of the learning process
x_iter = []
y_iter = []

momentum = 0

for m in range(number_of_iteration):
    
    error = 2*x_start+np.cos(2*x_start)*6
    
    momentum = lr*error+0.9*momentum # THE NEW ADDED LINE. Momentum
    
    x_start = x_start - momentum # de update rule
     
    x_iter += [x_start]
    y_iter += [x_start**2+np.sin(2*x_start)*3]
```

```python
x = np.linspace(-2, 5, num=100)
y = x**2+np.sin(2*x)*3

plt.axis([-2, 4, -3, 15])
plt.grid()
plt.plot(x,y)
plt.plot(x_iter,y_iter,'o',color='green')
plt.title('Gradient descent with Momentum. It finds the global minimum')
plt.show()
```
