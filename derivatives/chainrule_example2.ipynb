```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's say that we have a function $sin(x^{2})$. This is a $sin()$ function and inside is a quadratic function $x^{2}$. We take the derivative of the sine function and then multiply that with the derivative of the square. This results in $cos(x^{2})*2x$

```python
x = np.linspace(-5, 5, num=500)
y = np.sin(x**2)
```

```python
xi = 0.5
yi = np.sin(xi**2)
```

```python
a = np.cos(xi**2)*2*xi # the derivative!!!!

b = yi-a*xi
```

```python
tangent = a*x+b
```

See the new function using the chain rule. The tangent is as we expected. 

```python
plt.axis([-3, 3, -1.2, 1.2])
plt.grid()
plt.plot(x,y)
plt.plot(x,tangent,color='red')
plt.plot(xi,yi,'o',color='orange')
plt.show()
```
