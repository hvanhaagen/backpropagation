```python
import numpy as np
import matplotlib.pyplot as plt
```

We start by making a quadratic function

```python
x = np.linspace(-3, 3, num=100)
y = x**2
```

Then we choose a point on this quadratic function where we want to calculate the tangent for.

```python
xi = -1.5 # you can change this value to get another point
yi = xi**2
```

With the derivative we can calculate the slope of the tangent

```python
a = 2*xi # This is the derivative of the function. It defines the slope of the tangent

b = yi-a*xi # calculate the intercept once you know the slope
```

```python
tangent = a*x+b
```

Now we are going to plot the quadratic curve and the tangent (in red) in one figure.

```python
plt.axis([-3, 3, -1, 10])
plt.grid()
plt.plot(x,y)
plt.plot(x,tangent,color='red')
plt.plot(xi,yi,'o',color='orange')
plt.show()
```
