```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3, num=100)
y = np.sin(x)
```

```python
xi = 2
yi = np.sin(xi)
```

```python
a = np.cos(xi) # The derivative
b = yi-a*xi
```

```python
tangent = a*x+b
```

```python
plt.axis([-3, 3, -1.2, 1.2])
plt.grid()
plt.plot(x,y)
plt.plot(x,tangent,color='red')
plt.plot(xi,yi,'o',color='orange')
plt.show()
```
