```python
import numpy as np
import matplotlib.pyplot as plt
```

Let's say that we have a function $sin(x)^{2}$. This is a quadratic function $x^{2}$. And inside is another function the $sin(x)$. We take the derivative of the square and then multiply that with the derivative of the sin(x). This results in $2cos(x)$

```python
x = np.linspace(-3, 3, num=100)
y = np.sin(x)**2
```

```python
xi = 1
yi = np.sin(xi)**2
```

```python
a = 2*np.cos(xi) # the derivative!!!!
b = yi-a*xi
```

```python
tangent = a*x+b
```

See the new function using the chain rule. The tangent is as we expected. 

```python
plt.axis([-3, 3, -1.2, 1.2])
plt.grid()
plt.plot(x,y)
plt.plot(x,tangent,color='red')
plt.plot(xi,yi,'o',color='orange')
plt.show()
```
