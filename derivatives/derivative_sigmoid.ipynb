```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, num=100)
y = sigmoid(x)
```

```python
xi = 2.4
yi = sigmoid(xi)
```

```python
a = sigmoid(xi)*(1-sigmoid(xi)) # the derivative!!!!
b = yi-a*xi
```

```python
tangent = a*x+b
```

```python
plt.axis([-10, 10, -0.1, 1.1])
plt.grid()
plt.plot(x,y)
plt.plot(x,tangent,color='red')
plt.plot(xi,yi,'o',color='orange')
plt.show()
```
