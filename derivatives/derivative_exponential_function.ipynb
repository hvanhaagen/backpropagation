```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, num=100)
y = np.exp(x)
```

```python
xi = 0
yi = np.exp(xi)
```

```python
a = np.exp(xi) # The derivative!!!!
b = yi-a*xi
```

```python
tangent = a*x+b
```

```python
plt.axis([-5, 3, -0.1, 10])
plt.grid()
plt.plot(x,y)
plt.plot(x,tangent,color='red')
plt.plot(xi,yi,'o',color='orange')
plt.show()
```
