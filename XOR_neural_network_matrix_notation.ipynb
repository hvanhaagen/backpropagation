```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

# The XOR problem using a neural network

In this notebook we are going to discuss the XOR problem. This is a classical problem that shows the power of a neural network. We are going to use this great tutorial from youtube:  https://www.youtube.com/watch?v=kNPGXgzxoHw
Here the weights and biases are pre-calculated. 
I want to implement this and show you how it is done in the video, but also how to use **matrix notation**. The matrix notation becomes handy when the networks become bigger. 

```python
def sigmoid(x):
    y = 1/(1 + np.exp(-x))
    return y
```

First we are going to generate dummy data to visualise the XOR prediction,. 

```python
# dummy data
x1 = np.random.rand(1000)
x2 = np.random.rand(1000)
dataset = np.array([x1,x2])
dataset = np.transpose(dataset)
```

```python
# implement the  XOR neural net as discuss in the video. Link above
y = []
for X1 in dataset:
    h1 = sigmoid( 20*X1[0] + 20*X1[1] - 10)
    h2 = sigmoid(-20*X1[0] - 20*X1[1] + 30)

    z  = sigmoid( 20*h1 + 20*h2 - 30)
    y += [z]
```

```python
# Visualize the XOR neural net
ind1 = np.where(np.array(y) >0.5)  # classification. Map de output to labels
ind2 = np.where(np.array(y) <0.5)

plt.scatter(x1[ind1], x2[ind1],color="red")
plt.scatter(x1[ind2], x2[ind2],color="green")
plt.show()
```

Now we are going to do the same using **matrix notation**. This is a more compact way of implementing a neural network. Especially when we have a large neural net. 
For the backpropagation we also want to use matrix notation. 

```python
# All the weights are put in a matrix or vector. 
# Note that b2 is a scalar. Therefor lower case notation
W1 = np.array([[20,20],[-20,-20]])
B1 = np.array([-10,30])
b2 = -30
W2 = np.array([20,20])
```

```python
result = []
for X1 in dataset:
    
    Y1 = np.matmul(W1,X1)+B1
    X2 = sigmoid(Y1)
    Y2 = np.matmul(W2,X2)+b2
    out = sigmoid(Y2)
    result += [out]
```

```python
# Visualize the XOR neural net
ind1 = np. where(np.array(result) >0.5)  # classificatie. Map de output naar labels
ind2 = np. where(np.array(result) <0.5)

plt.scatter(x1[ind1], x2[ind1],color="red")
plt.scatter(x1[ind2], x2[ind2],color="green")
plt.show()
```
